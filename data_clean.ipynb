```python
# Concatenate the two DataFrames by rows and perform an inner join
joined_ASX_sector_data_rows = pd.concat(
    [SP_ASX_200_Utilities_Historical_price_Data_df, SP_ASX_200_Telecommunication_Services_Historical_price_Data_df, SP_ASX_200_Industrials_Historical_price_Data_df, 
     SP_ASX_200_Consumer_Discretionary_price_df, SP_ASX_200_Consumer_Staples_Historical_price_Data_df, SP_ASX_200_Energy_Historical_price_Data_df, 
     SP_ASX_200_REIT_Historical_Data_price_df, SP_ASX_200_Financials_Historical_Data_price_df, SP_ASX_200_Health_Care_Historical_Data_price_df, SP_ASX_200_Materials_Historical_price_Data_df,
     SP_ASX_200_Information_Technology_Historical_price_Data_df     
    ], axis=1, join="inner"
)
joined_ASX_sector_data_rows.head()
```

```python
#Amend column names to sector names
columns = ['Utilities','Telecommunication_Services','Industrials','Consumer_Discretionary','Consumer_Staple','Energy','Real Estate','Financials','Health_Care','Materials','Information_Technology']
joined_ASX_sector_data_rows.columns = columns
joined_ASX_sector_data_rows.head()
```
