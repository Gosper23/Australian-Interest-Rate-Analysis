```python
# Initial imports
import pandas as pd
from pathlib import Path
```

```python
# Create dataframe andClean ASX 200 historical data
csv_path = Path("data/S&P_ASX 200 Historical Data.csv")
SP_ASX_200_Historical_Data_df = pd.read_csv(csv_path, index_col="Date")
```

```python
# Check for null values
SP_ASX_200_Historical_Data_df.isnull()
```

```python
# Determine number of nulls
SP_ASX_200_Historical_Data_df.isnull().sum()
```

```python
# Check duplicates
SP_ASX_200_Historical_Data_df.duplicated()
```

```python
SP_ASX_200_Historical_Data_df.duplicated().sum()
```

```python
SP_ASX_200_Historical_Data_df.head()
```
